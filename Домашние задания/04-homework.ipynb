#### Hard
Написать запрос, который посчитает какую-то сложную аналитику по БД, используя как можно больше таблиц. Построить графики и сделать выводы.

Надо объединить регистрационный список (ФИО прошедших на тур, номер школы, балл за 1 тур) с данными ведомости 2 тура (рассчитать итоговый балл за 2 тур). Рассчитать по всем участникам итоговые баллы за два тура, распеределить грамоты и дипломы, исходя из того, что дипломантов (Победителей и Призеров) должно быть не более 35% от прошедших на 2 тур участников, а победителей не более 10%. Сделать аналитику по работе экзаменаторов.

```python
dt = pd.read_sql(
    """
    SELECT * FROM regspisok
    JOIN (
        select ШКОЛА as ШК, НАЗВАНИЕ_ШК from schools
        )schools on regspisok.ШКОЛА = schools.ШК
    LEFT JOIN (
        select РЕГНОМЕР_УЧ as РЕГ2, sum(БАЛЛ_2ТУР) AS ИТОГ_2ТУР from vedomost
        GROUP BY РЕГ2
        )vedomost on regspisok.РЕГНОМЕР_УЧ = vedomost.РЕГ2
    """,
    sql)
```

```python
data = dt[["РЕГНОМЕР_УЧ", "ФИО", "НАЗВАНИЕ_ШК", "БАЛЛ_1ТУР", "ИТОГ_2ТУР"]]
datad = data.fillna(0)
datad["ИТОГ_БАЛЛ"] = datad.БАЛЛ_1ТУР + datad.ИТОГ_2ТУР
dt_sorted = datad.sort_values(by="ИТОГ_БАЛЛ", ascending=False).reset_index(drop=True)
N = dt_sorted["РЕГНОМЕР_УЧ"].count() # число участников тура
W = int(N*0.1)                     # число Победителей
PW = int(N*0.35) - int(N*0.1)        # число Призеров
dt_sorted.insert(6, "РЕЗУЛЬТАТ", np.arange(N), False)
for x in dt_sorted.index:
    if (int(x) + 1) <= W:
        dt_sorted["РЕЗУЛЬТАТ"].loc[x] = "Победитель"
    elif (int(x) + 1) <= PW:
        dt_sorted["РЕЗУЛЬТАТ"].loc[x] = "Призер"
    else:
        dt_sorted["РЕЗУЛЬТАТ"].loc[x] = "Участник"
dt_sorted
```

```python
sns.histplot(dt_sorted["ИТОГ_БАЛЛ"], binwidth=15)
```

```python
sns.pairplot(dt_sorted[["БАЛЛ_1ТУР", "ИТОГ_2ТУР"]])
```

```python
dt_ex = pd.read_sql(
    """
    SELECT * FROM examenators
    JOIN stantions USING (ИД_СТАНЦИИ)
    JOIN(select ИД_ЭКЗАМ, avg(БАЛЛ_2ТУР) AS СР_БАЛЛ, min(БАЛЛ_2ТУР) AS МИН_БАЛЛ, max(БАЛЛ_2ТУР) AS МАКС_БАЛЛ from vedomost
    GROUP BY ИД_ЭКЗАМ) vedomost USING (ИД_ЭКЗАМ)
    """,
    sql)
```
